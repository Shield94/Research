```python
#Important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#plotting with taste
sns.set_theme()
sns.set_context("poster")
plt.style.use('fivethirtyeight')
```

```python
#Reading in the sets of data
s12 = pd.read_fwf("s12_o0_2D_gw2.dat")
s40 = pd.read_fwf("s40_o0_2D_gw2.dat")
```

```python
#values we need to subtract from the time column
s12_time_val = 0.184469701 
s40_time_val = 0.366373606

#Extracting the time values and adjusting them
s12_t = s12["#1 time"] - s12_time_val
s40_t = s40["#1 time"] - s40_time_val

#Solar Mass Conversion for later
solar_mass = 2e33 #g

#Conversion
s12_m = s12["2 mass"]/solar_mass
s40_m = s40["2 mass"]/solar_mass
```

```python
#Plotting Neutron Star Mass Vs Time
plt.figure(figsize = (15,7))

#s12_file
plt.subplot(121)
plt.plot(s12_t, s12_m)
plt.title("Neutron Star Mass vs Time")
plt.xlabel("Time [sec]")
plt.ylabel("Neutron Star Mass [$M_{\odot}$]")

#s40_file
plt.subplot(122)
plt.plot(s40_t, s40_m)
plt.title("Neutron Star Mass vs Time")
plt.xlabel("Time [sec]")
plt.ylabel("Neutron Star Mass [$M_{\odot}$]")

plt.tight_layout()
plt.show()
```

```python
#Unit Conversions
solar_radii = 6.957e+10 #cm

#Extracting Correct PNS Radius and adjusting
s12_r = s12["30 correct PNS radius"]/solar_radii
s40_r = s40["30 correct PNS radius"]/solar_radii
```

```python
#Plotting Correct PNS Radius Vs Time
plt.figure(figsize = (15,7))

#s12_file
plt.subplot(121)
plt.plot(s12_t, s12_r)
plt.title("Correct PNS Radius vs Time")
plt.xlabel("Time [sec]")
plt.ylabel("Correct PNS Radius [$R_{\odot}$]")

#s40_file
plt.subplot(122)
plt.plot(s40_t, s40_r)
plt.title("Correct PNS Radius vs Time")
plt.xlabel("Time [sec]")
plt.ylabel("Correct PNS Radius [$R_{\odot}$]")

plt.tight_layout()
plt.show()
```

```python
def derivative(func,time):
    derv = []
    t_new = []
    for i in range(0,len(func)-1):
        change_x = time[i+1] - time[i]
        change_y = (func[i+1] - func[i])

        derv.append((change_y)/(change_x))
        
        t_new.append((time[i+1] + time[i])/2)
        
    return [t_new,derv]
```

```python
#Define new times and gradients from function.
s12_t_new,s12_derv = derivative(s12_r,s12_t)
s40_t_new,s40_derv = derivative(s40_r,s40_t)
```

```python
#Plotting Correct PNS Radius Vs Time
plt.figure(figsize = (15,7))

#s12_file
plt.subplot(121)
plt.plot(s12_t, s12_r)
plt.plot(s12_t_new, s12_derv, lw = .1)
plt.title("Correct PNS Radius vs Time")
plt.xlabel("Time [sec]")
plt.ylabel("Correct PNS Radius [$R_{\odot}$]")

#s40_file
plt.subplot(122)
plt.plot(s40_t, s40_r)
plt.plot(s40_t_new, s40_derv, lw= .3)
plt.title("Correct PNS Radius vs Time")
plt.xlabel("Time [sec]")
plt.ylabel("Correct PNS Radius [$R_{\odot}$]")

plt.tight_layout()
plt.show()
```
